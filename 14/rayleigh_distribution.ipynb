```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 5))
for sigma in np.arange(0.5, 5, 0.5):
    x = np.linspace(sps.rayleigh(scale=sigma).ppf(0.00),
                    sps.rayleigh(scale=sigma).ppf(0.99), 100)
    plt.plot(x, sps.rayleigh(scale=sigma).pdf(x),
           '-', lw=2, alpha=0.6, label=f'$ \sigma = {sigma}$')
plt.grid()
plt.title('Rayleigh pdf')
plt.legend()
plt.show()


plt.figure(figsize=(15, 5))
for sigma in np.arange(0.5, 5, 0.5):
    x = np.linspace(sps.rayleigh(scale=sigma).ppf(0.00),
                    sps.rayleigh(scale=sigma).ppf(0.99), 100)
    plt.plot(x, sps.rayleigh(scale=sigma).cdf(x),
           '-', lw=2, alpha=0.6, label=f'$ \sigma = {sigma}$')
plt.grid()
plt.title('Rayleigh cdf')
plt.legend()
plt.show()
```
