$ (x^2 + y^2 - 1)^3 - x^2 y^3 = 0: $
    
$ y = \frac{1}{2} x^\frac{2}{3} \pm \sqrt{\frac{1}{4} x^\frac{4}{3} - x^2 + 1} $

```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
array = []
for i in np.arange(-1.2, 1.2, 0.001).round(3):
    x = abs(i)
    array.append([i, 1/2 * x**(2/3) + np.sqrt(1/4 * x**(4/3) - x**2 + 1)])
    array.append([i, 1/2 * x**(2/3) - np.sqrt(1/4 * x**(4/3) - x**2 + 1)])
array = np.array(array)
plt.plot(array[:, 0], array[:, 1], 'r-')
plt.show()
```
