```python
import requests as req
import json
from random import random
from tqdm import tqdm
from fake_headers import Headers
from bs4 import BeautifulSoup

req = req.Session()

headers = Headers(browser="chrome", os="win", headers=True).generate()

data = {'data': []}
for page in range(0, 99999):
    url = f'https://krasnoyarsk.hh.ru/search/vacancy?text=python+разработчик' \
           '&from=suggest_post&area=&page={page}&hhtmFrom=vacancy_search_list&items_on_page=50'   
    resp = req.get(url, headers=headers, allow_redirects=False)
    soup = BeautifulSoup(resp.text,"lxml")
    titles = soup.find_all(attrs={'data-qa':'serp-item__title'})
    
    if len(titles) == 0:
        break
        
    for i in tqdm(titles):
        url_object = i.attrs['href']
        resp_object = req.get(url_object, headers=headers)
        soup_object = BeautifulSoup(resp_object.text, 'lxml')
        
        try:
            exp = soup_object.find(
                attrs={"class":"vacancy-description-list-item"}).find(
                attrs={"data-qa":"vacancy-experience"}).text
        except:
            exp = '-'
        try:
            sal = soup_object.find(
                attrs={"data-qa":"vacancy-salary"}).find(
                attrs={"data-qa":"vacancy-salary-compensation-type-net"}).text
        except:
            sal = '-'
        try:
            adr = soup_object.find(
                attrs={"data-qa":"vacancy-address-with-map"}).find(
                attrs={"data-qa":"vacancy-view-raw-address"}).text
        except:
            adr = '-'
        
        data['data'].append({
            'title': i.text,
            'work experience': exp,
            'salary': sal,
            'region': adr.split(',')[0]
            })
        
        with open(f'data.json', 'w', encoding='utf-8') as file:
            json.dump(data, file, ensure_ascii=False)
    time.sleep(random()*10)
```
